```python
filtered = fullCoinData[fullCoinData['coin'] == COIN]
coin_df = pd.read_csv(f'fullData/{COIN}_df.csv')
coin_df['date'] = pd.to_datetime(coin_df['date'])
```

```python
all = pd.merge(coin_df, filtered, how="left", left_on='date', right_on='Created_at')
```
