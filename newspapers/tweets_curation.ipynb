```python
import pandas as pd
import numpy as np
from functions import *
```

```python
COIN = 'BTC'
```

```python
fullCoinData = pd.read_csv('fullCoinData.csv')
coin_df = pd.read_csv(f'fullData/{COIN}_df.csv')
fullCoinData['Created_at'] = pd.to_datetime(fullCoinData['Created_at'], format="%a %b %d %H:%M:%S %z %Y")
# Get the date part of the datetime
fullCoinData['date'] = fullCoinData['Created_at'].dt.date
fullCoinData['date'] = fullCoinData['date'].astype(str)

coin_df['date'] = pd.to_datetime(coin_df['date'])
coin_df['date'] = coin_df['date'].dt.date
coin_df['date'] = coin_df['date'].astype(str)
coin_df['time'] = coin_df['time'].astype(str)
```

```python
filtered = fullCoinData[fullCoinData['coin'] == COIN]
filtered['date'] = filtered['date'].astype(str)
```

```python
# Perform the merge with unique suffixes
all = pd.merge(coin_df, filtered, how="left", left_on='time', right_on='date', suffixes=('_coin', '_filtered'))
all
```

```python
all.to_csv(fullDataPath(COIN), index=False)
```
