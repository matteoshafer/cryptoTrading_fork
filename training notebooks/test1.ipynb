# Crypto Fear and Greed Index

```python
def fetch_fear_greed_index(url):
    """
    Fetch the current crypto Fear & Greed index from Alternative.me.
    Returns a dictionary with 'timestamp', 'source', 'value', and 'text' (as a string description).
    """
    response = requests.get(url, params={"limit": 1})
    if response.status_code != 200:
        raise Exception("Failed to fetch Fear & Greed data")
    data = response.json()
    if "data" in data and len(data["data"]) > 0:
        result = data["data"][0]
        # Convert the provided timestamp (in seconds) to a datetime
        ts = datetime.fromtimestamp(int(result["timestamp"]))
        # Construct a descriptive text based on the index value and classification.
        text = f"Fear & Greed Index: {result['value']} ({result['value_classification']})"
        return {
            "timestamp": ts,
            "source": "FearGreed",
            "value": result["value"],
            "text": text
        }
    return None

# Fetch the Fear & Greed index data and preview it
fear_greed_data = fetch_fear_greed_index(FEAR_GREED_URL)
df_fear_greed = pd.DataFrame([fear_greed_data])
print("Fear & Greed Data:")
display(df_fear_greed)
```
